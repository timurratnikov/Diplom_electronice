```python
import pandas as pd
import numpy as np

df = pd.read_excel('electronika_clean.xlsx')
```

```python
actions_values = df['Actions'].value_counts().index
actions_set = set()
for action in actions_values:
    values = action.split(',')
    for value in values:
        actions_set.add(value)

for col in list(actions_set):
    df[col] = df['Actions'].apply(lambda x: int(str(x).find(col) != -1))
df
```

```python
df[df['TN'] == 'Крупная бытовая техника']
```

```python
print(list(actions_set))
```

```python
# Выручка размещенные
action_list = list(actions_set)
category1 = df[df['TN'] == 'Крупная бытовая техника']
total = category1[['OrderDate', 'RowSum']].groupby('OrderDate').sum()

dfs = []

for col in action_list:
    stat = pd.DataFrame()
    stat['OrderDate'] = category1['OrderDate']
    stat[col] = category1[col] * category1['RowSum']
    dfs.append(stat.groupby('OrderDate').sum())

stat = pd.DataFrame()
stat['OrderDate'] = category1['OrderDate']
stat['БезАкций'] = category1['RowSum'] * (1 - category1[action_list[0]]) * (1 - category1[action_list[1]]) * (1 - category1[action_list[2]]) * (1 - category1[action_list[3]]) * (1 - category1[action_list[4]]) * (1 - category1[action_list[5]]) * (1 - category1[action_list[6]])
dfs.append(stat.groupby('OrderDate').sum())
```

```python
joined = total.copy()
for df1 in dfs:
    joined = joined.join(df1)

for col in joined.columns:
    if col == 'RowSum':
        continue
    joined[col] = 100.0 * joined[col] / joined['RowSum']

concatenated = pd.DataFrame()

for col in joined.columns:
    if col == 'RowSum':
        continue
    stat = joined[[col]].rename(columns={col: 'Percent'})
    stat['type'] = col
    concatenated = pd.concat([concatenated, stat])

concatenated = concatenated.reset_index()
concatenated
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.figure(figsize=(11, 6))
sns.lineplot(data=concatenated, x='OrderDate', y='Percent', hue='type')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title="Акция")
plt.title('Выручка (распределение акций)')
```

```python
# Спрос размещенные

action_list = list(actions_set)
category1 = df[df['TN'] == 'Крупная бытовая техника']
total = category1[['OrderDate', 'Quant']].groupby('OrderDate').sum()

dfs = []

for col in action_list:
    stat = pd.DataFrame()
    stat['OrderDate'] = category1['OrderDate']
    stat[col] = category1[col] * category1['Quant']
    dfs.append(stat.groupby('OrderDate').sum())

stat = pd.DataFrame()
stat['OrderDate'] = category1['OrderDate']
stat['БезАкций'] = category1['Quant'] * (1 - category1[action_list[0]]) * (1 - category1[action_list[1]]) * (1 - category1[action_list[2]]) * (1 - category1[action_list[3]]) * (1 - category1[action_list[4]]) * (1 - category1[action_list[5]]) * (1 - category1[action_list[6]])
dfs.append(stat.groupby('OrderDate').sum())

joined = total.copy()
for df1 in dfs:
    joined = joined.join(df1)

for col in joined.columns:
    if col == 'Quant':
        continue
    joined[col] = 100.0 * joined[col] / joined['Quant']

concatenated = pd.DataFrame()

for col in joined.columns:
    if col == 'Quant':
        continue
    stat = joined[[col]].rename(columns={col: 'Percent'})
    stat['type'] = col
    concatenated = pd.concat([concatenated, stat])

concatenated = concatenated.reset_index()
concatenated
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.figure(figsize=(11, 6))
sns.lineplot(data=concatenated, x='OrderDate', y='Percent', hue='type')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title="Акция")
plt.title('Quant (распределение акций)')
```
