```python
import pandas as pd
import numpy as np

df = pd.read_excel('electronika_clean.xlsx')
df
```

```python
df.info()
```

```python
all = df[['месяц', 'RowSum', 'Order_ID', 'Phone_new', 'Quant']]\
    .groupby(['месяц']) \
    .aggregate({'RowSum': 'sum', 'Order_ID': 'nunique', 'Phone_new': 'nunique', 'Quant': 'sum'})
all['Avg goods'] = (all['Quant'] / all['Order_ID']).apply(lambda x: round(x, 2))
all['Avg order'] = (all['RowSum'] / all['Order_ID']).apply(lambda x: round(x, 2))
all['Avg client'] = (all['RowSum'] / all['Phone_new']).apply(lambda x: round(x, 2))
all.rename(columns={'RowSum': 'Выручка, руб', 'Phone_new': 'Количество уникальных клиентов', 'Quant': 'Количество товаров', 'Order_ID': 'Количество чеков', 'Avg goods': 'Среднее количество', 'Avg order': 'Cредний чек, руб', 'Avg client': 'Ценность клиента'}, inplace=True)

total = df[['RowSum', 'Order_ID', 'Phone_new', 'Quant']] \
    .aggregate({'RowSum': 'sum', 'Order_ID': 'nunique', 'Phone_new': 'nunique', 'Quant': 'sum'})
total_dict = {
    'RowSum': [total.iloc[0]],
    'Order_ID': [total.iloc[1]],
    'Phone_new': [total.iloc[2]],
    'Quant': [total.iloc[3]]
}
total_df = pd.DataFrame().from_dict(total_dict)
total_df
total_df['Avg goods'] = (total_df['Quant'] / total_df['Order_ID']).apply(lambda x: round(x, 2))
total_df['Avg order'] = (total_df['RowSum'] / total_df['Order_ID']).apply(lambda x: round(x, 2))
total_df['Avg client'] = (total_df['RowSum'] / total_df['Phone_new']).apply(lambda x: round(x, 2))
total_df.rename(columns={'RowSum': 'Выручка, руб', 'Phone_new': 'Количество уникальных клиентов', 'Quant': 'Количество товаров', 'Order_ID': 'Количество чеков', 'Avg goods': 'Среднее количество', 'Avg order': 'Cредний чек, руб', 'Avg client': 'Ценность клиента'}, inplace=True)
all = pd.concat([all, total_df], axis=0)


sold = df[df['Status_ID'] == 14][['месяц', 'RowSum', 'Order_ID', 'Phone_new', 'Quant']]\
    .groupby(['месяц']) \
    .aggregate({'RowSum': 'sum', 'Order_ID': 'nunique', 'Phone_new': 'nunique', 'Quant': 'sum'})

sold['Avg goods'] = (sold['Quant'] / sold['Order_ID']).apply(lambda x: round(x, 2))
sold['Avg order'] = (sold['RowSum'] / sold['Order_ID']).apply(lambda x: round(x, 2))
sold['Avg client'] = (sold['RowSum'] / sold['Phone_new']).apply(lambda x: round(x, 2))
sold.rename(columns={'RowSum': 'Выручка, руб', 'Phone_new': 'Количество уникальных клиентов', 'Quant': 'Количество товаров', 'Order_ID': 'Количество чеков', 'Avg goods': 'Среднее количество', 'Avg order': 'Cредний чек, руб', 'Avg client': 'Ценность клиента'}, inplace=True)

total = df[df['Status_ID'] == 14][['RowSum', 'Order_ID', 'Phone_new', 'Quant']] \
    .aggregate({'RowSum': 'sum', 'Order_ID': 'nunique', 'Phone_new': 'nunique', 'Quant': 'sum'})
total_dict = {
    'RowSum': [total.iloc[0]],
    'Order_ID': [total.iloc[1]],
    'Phone_new': [total.iloc[2]],
    'Quant': [total.iloc[3]]
}
total_df = pd.DataFrame().from_dict(total_dict)
total_df
total_df['Avg goods'] = (total_df['Quant'] / total_df['Order_ID']).apply(lambda x: round(x, 2))
total_df['Avg order'] = (total_df['RowSum'] / total_df['Order_ID']).apply(lambda x: round(x, 2))
total_df['Avg client'] = (total_df['RowSum'] / total_df['Phone_new']).apply(lambda x: round(x, 2))
total_df.rename(columns={'RowSum': 'Выручка, руб', 'Phone_new': 'Количество уникальных клиентов', 'Quant': 'Количество товаров', 'Order_ID': 'Количество чеков', 'Avg goods': 'Среднее количество', 'Avg order': 'Cредний чек, руб', 'Avg client': 'Ценность клиента'}, inplace=True)
sold = pd.concat([sold, total_df], axis=0)
```

```python
all_by_category = df[['TN', 'RowSum', 'Order_ID', 'Phone_new', 'Quant']]\
    .groupby(['TN']) \
    .aggregate({'RowSum': 'sum', 'Order_ID': 'nunique', 'Phone_new': 'nunique', 'Quant': 'sum'})
all_by_category['Avg goods'] = (all_by_category['Quant'] / all_by_category['Order_ID']).apply(lambda x: round(x, 2))
all_by_category['Avg order'] = (all_by_category['RowSum'] / all_by_category['Order_ID']).apply(lambda x: round(x, 2))
all_by_category['Avg client'] = (all_by_category['RowSum'] / all_by_category['Phone_new']).apply(lambda x: round(x, 2))
all_by_category.sort_values(by='RowSum', ascending=False, inplace=True)
all_by_category.rename(columns={'RowSum': 'Выручка, руб', 'Phone_new': 'Количество уникальных клиентов', 'Quant': 'Количество товаров', 'Order_ID': 'Количество чеков', 'Avg goods': 'Среднее количество', 'Avg order': 'Cредний чек, руб', 'Avg client': 'Ценность клиента'}, inplace=True)

sold_by_category = df[df['Status_ID'] == 14][['TN', 'RowSum', 'Order_ID', 'Phone_new', 'Quant']]\
    .groupby(['TN']) \
    .aggregate({'RowSum': 'sum', 'Order_ID': 'nunique', 'Phone_new': 'nunique', 'Quant': 'sum'})

sold_by_category['Avg goods'] = (sold_by_category['Quant'] / sold_by_category['Order_ID']).apply(lambda x: round(x, 2))
sold_by_category['Avg order'] = (sold_by_category['RowSum'] / sold_by_category['Order_ID']).apply(lambda x: round(x, 2))
sold_by_category['Avg client'] = (sold_by_category['RowSum'] / sold_by_category['Phone_new']).apply(lambda x: round(x, 2))
sold_by_category.sort_values(by='RowSum', ascending=False, inplace=True)
sold_by_category.rename(columns={'RowSum': 'Выручка, руб', 'Phone_new': 'Количество уникальных клиентов', 'Quant': 'Количество товаров', 'Order_ID': 'Количество чеков', 'Avg goods': 'Среднее количество', 'Avg order': 'Cредний чек, руб', 'Avg client': 'Ценность клиента'}, inplace=True)
```

```python
join = all_by_category.join(sold_by_category, how='left', rsuffix='_sold')
join['Доля выкупа (выручка)'] = (100.0 * join['Выручка, руб_sold'] / join['Выручка, руб']).apply(lambda x : round(x, 2))
join['Доля выкупа (количество товаров)'] = (100.0 * join['Количество товаров_sold'] / join['Количество товаров']).apply(lambda x : round(x, 2))
```

```python
all_by_category = df[['TK', 'RowSum', 'Order_ID', 'Phone_new', 'Quant']]\
    .groupby(['TK']) \
    .aggregate({'RowSum': 'sum', 'Order_ID': 'nunique', 'Phone_new': 'nunique', 'Quant': 'sum'})
all_by_category['Avg goods'] = (all_by_category['Quant'] / all_by_category['Order_ID']).apply(lambda x: round(x, 2))
all_by_category['Avg order'] = (all_by_category['RowSum'] / all_by_category['Order_ID']).apply(lambda x: round(x, 2))
all_by_category['Avg client'] = (all_by_category['RowSum'] / all_by_category['Phone_new']).apply(lambda x: round(x, 2))
all_by_category.sort_values(by='RowSum', ascending=False, inplace=True)

sold_by_category = df[df['Status_ID'] == 14][['TK', 'RowSum', 'Order_ID', 'Phone_new', 'Quant']]\
    .groupby(['TK']) \
    .aggregate({'RowSum': 'sum', 'Order_ID': 'nunique', 'Phone_new': 'nunique', 'Quant': 'sum'})

sold_by_category['Avg goods'] = (sold_by_category['Quant'] / sold_by_category['Order_ID']).apply(lambda x: round(x, 2))
sold_by_category['Avg order'] = (sold_by_category['RowSum'] / sold_by_category['Order_ID']).apply(lambda x: round(x, 2))
sold_by_category['Avg client'] = (sold_by_category['RowSum'] / sold_by_category['Phone_new']).apply(lambda x: round(x, 2))
sold_by_category.sort_values(by='RowSum', ascending=False, inplace=True)

join = all_by_category.join(sold_by_category, how='left', rsuffix='_sold')
join['Доля выкупа (выручка)'] = (100.0 * join['RowSum_sold'] / join['RowSum']).apply(lambda x : round(x, 2))
join['Доля выкупа (количество товаров)'] = (100.0 * join['Quant_sold'] / join['Quant']).apply(lambda x : round(x, 2))
join.sort_values(by='RowSum', ascending=False, inplace=True)
join.rename(
    columns={
        'RowSum': 'Revenue total',
        'Order_ID': 'Orders total',
        'Phone_new': 'Clients total',
        'Quant': 'Goods total',
        'Avg order': 'Average bill total',
        'RowSum_sold': 'Revenue redeemed',
        'Доля выкупа (выручка)': 'Revenue redeemability',
        'Quant_sold': 'Goods redeemed',
        'Доля выкупа (количество товаров)': 'Goods redeemability'
    },
    inplace=True
)

join.head(20)
```
