```python
fs1 = df[['Phone_new', 'Order_ID', 'Quant', 'FullSum']] \
        .groupby(['Phone_new']) \
        .aggregate({'Order_ID': 'nunique', 'Quant': 'sum', 'FullSum': 'sum'}) \
        .rename(columns={'Order_ID': 'Количество чеков', 'Quant': 'Количество товаров', 'FullSum': 'Выручка'})
fs1['Среднее количество товаров в чеке'] = (fs1['Количество товаров'] / fs1['Количество чеков']).apply(lambda x: round(x, 2))
fs1
```

```python
fs2 = df[df['Status_ID'] == 14][['Phone_new', 'Order_ID', 'Quant', 'FullSum']] \
        .groupby(['Phone_new']) \
        .aggregate({'Order_ID': 'nunique', 'Quant': 'sum', 'FullSum': 'sum'}) \
        .rename(columns={'Order_ID': 'Количество выкупленных чеков', 'Quant': 'Количество выкупленных товаров', 'FullSum': 'Выручка выкупленных товаров'})
fs2['Среднее количество товаров в выкупленных чеках'] = (fs2['Количество выкупленных товаров'] / fs2['Количество выкупленных чеков']).apply(lambda x: round(x, 2))
fs2
```

```python
fs3 = fs1.join(fs2, how='left')
fs3.fillna(0, inplace=True)
fs3
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt

distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 16)
  
for k in K:
    # Building and fitting the model
    kmeans = KMeans(n_clusters=k)
    kmeanModel = kmeans.fit(fs3)
    kmeanModel.fit(fs3)
    distortions.append(sum(np.min(cdist(fs3, kmeanModel.cluster_centers_,
                                        'euclidean'), axis=1)) / fs3.shape[0])
    inertias.append(kmeanModel.inertia_)
  
    mapping1[k] = sum(np.min(cdist(fs3, kmeanModel.cluster_centers_,
                                   'euclidean'), axis=1)) / fs3.shape[0]
    mapping2[k] = kmeanModel.inertia_
```

```python
for key, val in mapping1.items():
    print(f'{key} : {val}')
```

```python
plt.plot(K, distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

```python
kmeans = KMeans(n_clusters=7)
kmeans.fit(fs3)
print(kmeans.cluster_centers_)
```

```python
labels = np.array(kmeans.labels_)
indexes = np.arange(len(labels))
clusters = []
for i in range(7):
    clusters.append(fs3.iloc[indexes[labels == i]])
```
